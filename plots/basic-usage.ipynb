```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
iris_table = pd.DataFrame(iris["data"],columns=iris['feature_names'])
iris_table["species"] = iris['target_names'][iris['target']]
```

## scatter plot

```python
fig, ax = plt.subplots(figsize=(5,4))
ax.scatter(iris_table["sepal length (cm)"].values,iris_table["petal length (cm)"].values,label="flowers")
ax.set_xlabel("sepal length (cm)")
ax.set_ylabel("petal length (cm)")
plt.legend()
plt.savefig("scatter-default.pdf",bbox_inches="tight")
```

```python
fig, ax = plt.subplots(figsize=(5,4))
ax = sns.scatterplot(data=iris_table,x="sepal length (cm)",y="petal length (cm)",hue="species",ax=ax)
plt.savefig("scatter-seaborn-default.pdf",bbox_inches="tight")
```

## bar chart

```python
np.random.seed(666)
h = np.random.randn(5)
h = h-h.min()+2
x = np.arange(5)
fig, ax = plt.subplots(figsize=(5,4))
ax.bar(x=x,height=h)
```

```python
fig, ax = plt.subplots(figsize=(5,4))
ax = sns.barplot(data=iris_table,x="species",y="petal length (cm)",ax=ax)
plt.savefig("bar-chart-seaborn-default.pdf",bbox_inches="tight")
```

## box plot, violin plot and strip plot

```python
fig, ax = plt.subplots(figsize=(5,4))
ax = sns.boxplot(data=iris_table,x="species",y="petal length (cm)",ax=ax)
plt.savefig("box-plot-seaborn-default.pdf",bbox_inches="tight")
```

```python
fig, ax = plt.subplots(figsize=(5,4))
ax = sns.violinplot(data=iris_table,x="species",y="petal length (cm)",ax=ax)
plt.savefig("violin-plot-seaborn-default.pdf",bbox_inches="tight")
```

```python
fig, ax = plt.subplots(figsize=(5,4))
ax = sns.stripplot(data=iris_table,x="species",y="petal length (cm)",ax=ax)
plt.savefig("strip-plot-seaborn-default.pdf",bbox_inches="tight")
```

```python
np.random.seed(666)
X = np.random.randn(50,10)
X[:25,5:] += 2
X[25:,5:] -= 2
Z = (X-X.mean(axis=1,keepdims=True))/X.std(axis=1,keepdims=True)
fig, ax = plt.subplots(figsize=(3,15))
cax = ax.matshow(Z)
fig.colorbar(cax)
plt.savefig("heatmap-default.pdf",bbox_inches="tight")
```
