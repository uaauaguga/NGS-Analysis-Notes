```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
iris_table = pd.DataFrame(iris["data"],columns=iris['feature_names'])
iris_table["species"] = iris['target_names'][iris['target']]
```

## scatter plot

```python
fig, ax = plt.subplots(figsize=(5,4))
ax.scatter(iris_table["sepal length (cm)"].values,iris_table["petal length (cm)"].values,label="flowers")
ax.set_xlabel("sepal length (cm)",fontsize=18,weight="bold")
ax.set_ylabel("petal length (cm)",fontsize=18,weight="bold")
ax.tick_params(labelsize=14)
plt.legend(fontsize=14)
```

```python
fig, ax = plt.subplots(figsize=(5,4))
ax = sns.scatterplot(data=iris_table,x="sepal length (cm)",y="petal length (cm)",hue="species",palette=["#C44E52","#CCB974","#64B5CD"],ax=ax)
plt.setp(ax.xaxis.label,fontsize=18,weight="bold")
plt.setp(ax.yaxis.label,fontsize=18,weight="bold")
ax.tick_params(labelsize=14)
plt.setp(ax.get_legend().get_texts(), fontsize=14) 
plt.setp(ax.get_legend().get_title(), fontsize=16,weight="bold") 
```

```python
np.random.seed(666)
h = np.random.randn(5)
h = h-h.min()+2
x = np.arange(5)
fig, ax = plt.subplots(figsize=(5,4))
ax.bar(x=x,height=h,edgecolor="black",width=0.5)
ax.tick_params(labelsize=14)
```

```python
fig, ax = plt.subplots(figsize=(4,4))
ax = sns.barplot(data=iris_table,x="species",y="petal length (cm)",
                 ax=ax,edgecolor="black",
                 palette=["#C44E52","#CCB974","#64B5CD"],linewidth=1.2)
# see https://stackoverflow.com/questions/34888058/changing-width-of-bars-in-bar-chart-created-using-seaborn-factorplot
width = 0.4
for patch in ax.patches :
    current_width = patch.get_width()
    diff = current_width - width
    # we change the bar width
    patch.set_width(width)
    # we recenter the bar
    patch.set_x(patch.get_x() + diff * .5)
    
plt.setp(ax.xaxis.label,fontsize=18,weight="bold")
plt.setp(ax.yaxis.label,fontsize=18,weight="bold")
_ = plt.setp(ax.get_xticklabels(),fontsize=16,rotation=45,ha="right")
_ = plt.setp(ax.get_yticklabels(),fontsize=16)
```

```python
fig, ax = plt.subplots(figsize=(4,4))
ax = sns.boxplot(data=iris_table,x="species",y="petal length (cm)",ax=ax,
                 showfliers=False,
                 boxprops={"edgecolor":"gray"},
                 palette=["#C44E52","#CCB974","#64B5CD"],width=0.4)
plt.setp(ax.xaxis.label,fontsize=18,weight="bold")
plt.setp(ax.yaxis.label,fontsize=18,weight="bold")
_ = plt.setp(ax.get_xticklabels(),fontsize=16,rotation=45,ha="right")
_ = plt.setp(ax.get_yticklabels(),fontsize=16)
```

```python
np.random.seed(666)
X = np.random.randn(50,10)
X[:25,5:] += 2
X[25:,5:] -= 2
Z = (X-X.mean(axis=1,keepdims=True))/X.std(axis=1,keepdims=True)

fig, ax = plt.subplots(figsize=(3,15))
ax = sns.heatmap(Z,cmap="coolwarm",linewidths=0.4,linecolor="gray",square=True,cbar_kws={"shrink":0.3})
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.tick_params(length=0)
ax.collections[0].colorbar.outline.set_edgecolor("gray")
ax.collections[0].colorbar.outline.set_linewidth(0.5)
plt.savefig("heatmap-seaborn.pdf",bbox_inches="tight")
```
